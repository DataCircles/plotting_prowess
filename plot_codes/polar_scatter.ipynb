# Polar Scatter Plot

The source: https://matplotlib.org/3.1.1/gallery/pie_and_polar_charts/polar_scatter.html

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np

from colors import Color_Coded
```

```python
cc = Color_Coded()
cmap = cc.color_map()
```

The following code is a reproduction of the matplotlib documentation, except for the color scheme, random seed, and rtick labels.

I used their example to learn the basics of polar scatter plots. The commentary was added for fun, and to suggest ways to better understand the different parameters used. 

```python
# Random seed is my childhood soccer number
np.random.seed(33)

# Set areas and colors of scattered circles
'''

'N' determines the upper limit of scattered circle radii. 

'r' is an array of random radii for the scattered circles. 

'theta' will determine both the location of the scattered circles and their colors.

'area' is an array of areas for the scattered circles, calculated with the 'r' array. 

'colors' is an array of numbers (theta) that determine a position along our colormap

---

Try out different values of N, or see what happens if you set 'colors' equal to 'r'. 

'''

N = 120
r = 2 * np.random.rand(N) 
theta = 2 * np.pi * np.random.rand(N)
area = 200 * r**2
colors = theta


fig = plt.figure()
ax = fig.add_subplot(111, polar=True)
c = ax.scatter(theta, r, c=colors, s=area, cmap=cmap, alpha=0.75)

# Optional: Change the radial origin. Try some other values : -10, -5, -0.5, 0, 1, 1.5
ax.set_rorigin(-2.5)
# Optional: Choose location of 0 degree mark. Try skipping this line. 
ax.set_theta_zero_location('W', offset=10)
# Optional: Hide radial labels. Try skipping this line. 
ax.set_rticks([])

plt.show()
```
