# Styling Subplots

A story of subplots: You know how to make subplots, and you find them quite handy. Maybe you're new to data science, or maybe your experience simply has not required the tedium of perfecting plots. You don't necessarily need something fancy, but you want to achieve a reasonably legible figure. The adjustment you hope to make can be done in Excel with one click, so it should be a quick fix... Right?

Alas, your Google searches yield dozens of Stack Overflow results, most of which are vaguely connected to the adjustment you have in mind, but not quite right. Matplotlib.pyplots has just enough of its own lingo to make your general plotting terminology inaccurate or imprecise as search terms. I personally love reading matplotlib documentation, but I still have a hard time finding the pertinent sections if I don't know the precise terms to look for.

This tutorial is not intended to teach you to be a master of matplotlib. (If you happen to have such a tutorial, please share it!!!) This is a sampler of "quick fixes" for making your subplots easier to understand and/or more pleasing to view. 

Some other subplot resources:

https://matplotlib.org/3.1.1/gallery/subplots_axes_and_figures/subplots_demo.html

````python
import matplotlib.pyplot as plt
import numpy as np

from colors import Color_Coded
cc = Color_Coded()
cmap = cc.color_map()

x = np.random.randint(1,50,75)
y = np.random.randint(1,50,75)

w = np.linspace(0,10)
z = np.linspace(5,15)
````

## Labels
### <span style="color:#5AADA4">Plot 1</span>
The most frustrating points I ever lost on my science homework were deducted for failing to label my axes. So let's start here.

#### Methods

* ax.set_title()
* ax.set_xlabel()
* ax.set_ylabel()

````python
fig, ax = plt.subplots(figsize=(4,4))

ax.scatter(x, y, c='#E85F69')
ax.set_title('Time for a Title')
ax.set_xlabel('Some Random Integers')
ax.set_ylabel('Some Random Integers')

plt.show()
````

<span style="color:#BD4D55"> Rating:  Bare Necessities. </span>
* No points lost for missing labels
* Not *unpleasant* to look at
* It is worth noting that this approach gets worse as you add subplots. We'll get to that below. 

### <span style="color:#5AADA4">Plot 2</span>

#### Methods
* ax.set_title()
* ax.set_xlabel()
* ax.set_ylabel()

##### Parameters
* change text size with **size**
* make bold text with **fontweight**
* adjust label spacing with **pad**

````python
fig, ax = plt.subplots(figsize=(4,4))

'''
While not part of this particular lesson, I will opt to use 
parameters 's', 'c', 'cmap', and 'alpha' to make my plots prettier.
Plus, I like the Data Circles cmap. 
'''
ax.scatter(x, y, s=200, c=x, cmap=cmap, alpha=0.7)
ax.set_title('Time for a Title', size=14, fontweight='bold', pad=10)
ax.set_xlabel('Some Random Integers')
ax.set_ylabel('Some Random Integers')

plt.show()
````

<span style="color:#BD4D55"> Rating:  A Semblance of Style. </span>

These aren't vast improvements, but they offer adaptability. If you care about the appearance, then your plot is likely intended to be displayed somehow. Whether this will be viewed in a notebook on a desktop or in a powerpoint on a large screen, you should probably have a preference about font size for your purposes. 

## Multiples

It's pretty easy to create 2 + subplots, but the automatic layout gets crowded. 

### Plot 1A

````python

fig, ax = plt.subplots(1, 2, figsize=(8,4))

ax[0].scatter(x, y, c='#E85F69')
ax[0].set_title('Tale of Two Titles')
ax[0].set_xlabel('Some Random Integers')
ax[0].set_ylabel('Some Random Integers')

ax[1].scatter(x, y, c='#E85F69')
ax[1].set_title('Tale of Two Titles')
ax[1].set_xlabel('Some Random Integers')
ax[1].set_ylabel('Some Random Integers')

fig.suptitle('The Super Title')

plt.show()
````

````python
fig, ax = plt.subplots(2, 2, figsize=(8,8))

ax[0,0].scatter(x, y, c='#E85F69')
ax[0,0].set_title('Time for Two Titles')
ax[0,0].set_xlabel('Some Random Integers')
ax[0,0].set_ylabel('Some Random Integers')

ax[1,0].scatter(x, y, c='#E85F69')
ax[1,0].set_title('Time for Two Titles')
ax[1,0].set_xlabel('Some Random Integers')
ax[1,0].set_ylabel('Some Random Integers')

ax[0,1].scatter(x, y, c='#E85F69')
ax[0,1].set_title('Time for Two Titles')
ax[0,1].set_xlabel('Some Random Integers')
ax[0,1].set_ylabel('Some Random Integers')
ax[1,1].scatter(x, y, c='#E85F69')
ax[1,1].set_title('Time for Two Titles')
ax[1,1].set_xlabel('Some Random Integers')
ax[1,1].set_ylabel('Some Random Integers')

fig.suptitle('The Super Title')

plt.show()
````

### Plot 2A

````python
fig, ax = plt.subplots(1, 2, figsize=(11,4), gridspec_kw={'wspace': .25})

ax[0].scatter(x, y, s=200, c=x, cmap=cmap, alpha=0.7)
ax[0].set_title('Time for Two Titles', size=14)
ax[0].set_xlabel('Some Random Integers')
ax[0].set_ylabel('Some Random Integers')

ax[1].scatter(x, y*7, s=200, c=x, cmap=cmap, alpha=0.7)
ax[1].set_title('Time for Two Titles', size=14)
ax[1].set_xlabel('Some Random Integers')
ax[1].set_ylabel('Some Random Integers')

fig.suptitle('The Super Title', fontsize=14, fontweight='bold', y=1.07)

plt.show()
````

### Plot 2B

````python
fig, axs = plt.subplots(2, 2, figsize=(11,11), gridspec_kw={'wspace': .25, 'hspace':.27})

for idx, ax in enumerate(axs.flatten()):
    ax.scatter(x, y, s=200, c=x, cmap=cmap, alpha=0.7)
    ax.set_title('Time for Two Titles', size=14)
    ax.set_xlabel('Some Random Integers')
    ax.set_ylabel('Some Random Integers')

fig.suptitle('Different Code for the Same Plot', fontsize=14, fontweight='bold', y=.95)

plt.show()
````

While the code used above is my preference, here is a version that better follows the style of previous examples. The resulting plots are indentical.

### Sharing Axes
is sometimes a good idea...

````python
fig, ax = plt.subplots(1, 2, figsize=(8,4), sharey=True)

ax[0].scatter(x, y, s=200, c=x, cmap=cmap, alpha=0.7)
ax[0].set_title('Time for Two Titles', size=14)
ax[0].set_xlabel('Some Random Integers')
ax[0].set_ylabel('Some Random Integers')

ax[1].scatter(x, y, s=200, c=x, cmap=cmap, alpha=0.7)
ax[1].set_title('Time for Two Titles', size=14)
ax[1].set_xlabel('Some Random Integers')

fig.suptitle('To Share a Y-Axis', fontsize=14, fontweight='bold', y=1.07)
fig.tight_layout()

plt.show()
````

````python
fig, ax = plt.subplots(1, 2, figsize=(8,4), sharey=True)

ax[0].scatter(x, y, s=200, c=x, cmap=cmap, alpha=0.7)
ax[0].set_title('Time for Two Titles', size=14)
ax[0].set_xlabel('Some Random Integers')
ax[0].set_ylabel('Some Random Integers')

ax[1].scatter(x, y*7, s=200, c=x, cmap=cmap, alpha=0.7)
ax[1].set_title('Time for Two Titles', size=14)
ax[1].set_xlabel('Some Random Integers')

fig.suptitle('Or Not to Share a Y-Axis', fontsize=14, fontweight='bold', y=1.07)
fig.tight_layout()

plt.show()
````

### Flipping Axes


````python
fig, ax = plt.subplots(1, 2, figsize=(10,4), gridspec_kw={'wspace': .2})

ax[0].scatter(x, y, s=200, c=x, cmap=cmap, alpha=0.7)
ax[0].set_title('Time for Two Titles', size=14)
ax[0].set_xlabel('Some Random Integers')
ax[0].set_ylabel('Some Random Integers')

ax[1].plot(w, z, c='#185675', alpha=0.7)
ax[1].set_title('Time for Two Titles', size=14)
ax[1].set_xlabel('Some Random Integers')
ax[1].set_ylabel('Some Random Integers', rotation=270, labelpad=20)
ax[1].yaxis.set_label_position("right")
ax[1].yaxis.tick_right()

fig.suptitle('The Super Title', fontsize=14, fontweight='bold', y=1.07)

plt.show()
````

````python
fig, ax = plt.subplots(1, 2, figsize=(8,4))

ax[0].scatter(x, y, s=200, c=x, cmap=cmap, alpha=0.7)
ax[0].set_title('Time for Two Titles', size=14)
ax[0].set_xlabel('Some Random Integers')
ax[0].set_ylabel('Some Random Integers')

ax[0].tick_params(axis='y', colors='#7E446C')
ax[0].yaxis.label.set_color('#7E446C')

ax[1].scatter(x, y*7, s=200, c=x, cmap=cmap, alpha=0.7)
ax[1].set_title('Time for Two Titles', size=14)
ax[1].set_xlabel('Some Random Integers')
ax[1].set_ylabel('Some Random Integers')
ax[1].yaxis.set_label_position("right")
ax[1].yaxis.tick_right()

ax[1].tick_params(axis='y', colors='#185675')
ax[1].yaxis.label.set_color('#185675')

fig.suptitle('The Super Title', fontsize=14, fontweight='bold', y=1.07)
fig.tight_layout()

plt.show()
````
