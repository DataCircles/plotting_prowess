```python
'''
For consistency, this series of blog & social media posts will always feature 
plots that use some part of our color scheme. We've created a few easy ways to use our colors. 
When you import the 'Color_Coded' class, you can use a pre-configured cmap. If you need a different
format, you can load the colors in dictionaries with color names and hex color codes. 

'''
from colors import Color_Coded

cc = Color_Coded()

cmap = cc.color_map()

# Dictionary with all theme colors
color_dict = cc.dct()

# Specify theme for a subset of colors: 'dark', 'medium', 'light'
dark_dict = cc.dct(theme='dark')
```

#### Using individual colors

```python
x = np.random.randint(1, 20, 40)
y = np.random.randint(1, 20, 40)

x2 = np.random.randint(1, 20, 40)
y2 = np.random.randint(1, 20, 40)

plt.scatter(x, y, c=color_dict['purple'], s=70)
plt.scatter(x2, y2, c=color_dict['blue'], s=70)
plt.show()
```

#### Using cmap

```python
x = np.random.randint(1, 50, 100)
y = np.random.randint(1, 50, 100)
c = np.random.randint(1, 50, 100)

plt.scatter(x, y, c=c, cmap=cmap, s=100)
```
