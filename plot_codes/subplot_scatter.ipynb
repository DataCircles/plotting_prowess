# Subplot Scatter

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
from colors import Color_Coded
cc = Color_Coded()
cmap = cc.color_map()
```

```python
fig, axs = plt.subplots(2, 2, figsize=(8,8), sharey=True, sharex=True)
np.random.seed(2020)

subtitles = ['Each subplot', 'has random data', 'and', 'pretty colors!']

for idx, ax in enumerate(axs.flatten()):
    x = np.random.randint(1,50,20)
    y = np.random.randint(1,50,20)
    z = x + np.random.randint(1,50,20)

    ax.scatter(x, y, s=z**2, c=x, cmap=cmap, alpha=0.7)
    ax.set_title(subtitles[idx])
    
    if idx % 2 == 0:
        ax.set_ylabel('Some Random Integers')
    if idx > 1:
        ax.set_xlabel('More Random Integers')

fig.suptitle('Subplot Scatter', fontsize=14, fontweight='bold', y=1.03)
fig.tight_layout()
```
